# Breast Cancer Data Set
         Breast cancer starts when cells in the breast begin to grow out of control.These cells usually form tumors that can be seen via X-ray or felt as  lumps in the breast area.
         
 Attribute Information
 
 Diagnosis:Malignant(M)(Cancerous),Benign(B)(Non Cancerous
 
 Radius mean:mean radius of the cell nuclei
 
 Texture mean:Variation in pixel intensity and patterns in the range
 
 Perimeter mean: mean perimeter of the cell nuclei
 
 Area mean:area perimeter of the cell nuclei
 
 Smoothness mean:measure of variation in pixel intensity /gray levels within the nuclei,describes the regularity/irregularity of the nuclei
 
 Compact mean: measur that quantities closely the boundaries of the cell nuclei adheres to the actual area
 
 concavity mean: measure the degree of inward and outward curvature of the cell nuclei's boundary
 
 concave points mean : mean bo. of concave points in the cells nuclei
 
 symmetry mean: mean symmetry of cell nuclei in the breast mass image,irregularities in cell symmetry may be associated with mallignancy.
 
 Fractal dimension  mean:mathematical measure that can describe the complexity and irregularities of shapes within the image 
 radius_se:
Standard error of the mean distance from the center to points on the perimeter of the cell nuclei.

texture_se:
Standard error of the standard deviation of gray-scale values.

perimeter_se:
Standard error associated with the perimeter measurements of the cell nuclei.

area_se:
Standard error of the area occupied by the cell nuclei.

smoothness_se:
Standard error for the local variation in radius lengths.

compactness_se:
Standard error for the compactness measurement (perimeter^2/area - 1.0).

concavity_se:
Standard error for the severity of concave portions of the contour.

concave_points_se:
Standard error for the number of concave portions of the contour.

symmetry_se:
Standard error for the symmetry of cell nuclei.

fractal_dimension_se:
Standard error for the "coastline approximation" - 1.
                       


radius_worst:
Mean of the three largest distances from the center to points on the perimeter of cell nuclei.

texture_worst:
Mean of the three largest standard deviations of gray-scale values. This provides information about the variations in the grayscale texture of the most irregular cells.

perimeter_worst:
Mean of the three largest perimeters of cell nuclei. This value can indicate the irregularity of the shape of the cell nuclei.

area_worst:
Mean of the three largest areas occupied by cell nuclei, providing insight into the size of the largest nuclei.

smoothness_worst:
Mean of the three largest local variations in radius lengths, offering a sense of the roughness or irregularity in the outer boundary of the largest nuclei.
   
compactness_worst:
Mean of the three largest compactness measurements, which is a ratio relating the perimeter and area of the cell (perimeter^2/area - 1.0). It provides a sense of how compact the largest cell nuclei are.

concavity_worst:
Mean of the three largest severities of concave portions of the contour. This indicates the extent of concavities in the largest cell nuclei.

concave_points_worst:
Mean of the three largest numbers of concave portions of the contour, indicating how many concave points appear in the most irregular cell nuclei.

symmetry_worst:
Mean of the three largest symmetry values of the cell nuclei, offering insight into how asymmetrical the largest nuclei are.

fractal_dimension_worst:
Mean of the three largest "coastline approximation" values minus one. It quantifies how much the boundary of the largest cell nuclei resembles a fractal.
